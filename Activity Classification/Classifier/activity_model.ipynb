```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.svm import SVC
import pickle
```

### Splitting into training and testing dataset

```python
data = pd.read_csv('dataset/train_dataset.csv')
x_train = data.iloc[:,:-1].values
y_train = data.iloc[:,-1].values
```

```python
data = pd.read_csv('dataset/train_dataset.csv')
x_test= data.iloc[:,:-1].values
y_test = data.iloc[:,-1].values
```

```python
from sklearn.linear_model import LogisticRegression    #supervised model (takes input+output find relationship)
lr=LogisticRegression(solver='sag')
lr.fit(x_train,y_train)
y_pre1=lr.predict(x_test)

#generating accuracy score and construction confusion matrix

from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy', accuracy_score(y_pre1, y_test)*100)
import seaborn as sb
plt.rcParams.update({'font.size': 10})
plt.figure(figsize = (5,5))

labels = ['dancing', 'running', 'sitting','standing']
cm=confusion_matrix(y_test,y_pre1, labels)
sb.heatmap(cm,annot=True, cmap="OrRd", fmt=".0f")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
# knn=KNeighborsClassifier(n_neighbors=5)
knn=KNeighborsClassifier(n_neighbors=5, p=1)
knn.fit(x_train,y_train)
y_pre2=knn.predict(x_test)
print(accuracy_score(y_test,y_pre2)*100)
plt.rcParams.update({'font.size': 24})
plt.figure(figsize = (12,12))
cm=confusion_matrix(y_test,y_pre2)
import seaborn as sb
sb.heatmap(cm,annot=True, cmap="OrRd", fmt=".0f")
```

# SVM

```python
from sklearn.svm import SVC
svm1=SVC(kernel='linear',)
svm1.fit(x_train,y_train)
y_pre3=svm1.predict(x_test)
print(accuracy_score(y_test,y_pre3)*100)
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pre3)
import seaborn as sb
sb.heatmap(cm,annot=True, cmap="OrRd", fmt=".0f")
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=6)
dt.fit(x_train,y_train)
y_pre5=dt.predict(x_test)
print(accuracy_score(y_test,y_pre5)*100)
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pre5)
from sklearn.utils.multiclass import unique_labels
unique_labels(y_test, y_pre5)
import seaborn as sb
sb.heatmap(cm,annot=True, cmap="OrRd", fmt=".0f")
```
