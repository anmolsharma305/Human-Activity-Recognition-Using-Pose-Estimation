```python
from sklearn import svm
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pose = pd.read_csv('dataset/multiclass.csv')
```

```python
pose
```

```python
print(f'Numbers of row/examples and column in dataset: {pose.shape}')
```

```python
# summary of information on the dataset
pose.info()
```

### Cleaning the data


```python
print("Display NA values in each columns: ")
pose.isna().sum(axis=0)
```

```python
print("Dispaly Null Values in each columns: ")
pose.isnull().sum()
```

```python
pose = pose.dropna()
```

```python
# shuffle the data row-wise
pose = pose.sample(frac=1)
```

```python
pose
```

### Split the data into train/test splits

```python
#split the data into 60% train and 40% test (later will divide the test to test and validate)
train_dataset, temp_test_dataset = train_test_split(pose, test_size=0.4)
```

```python
print(train_dataset.shape)
print(temp_test_dataset.shape)
```

```python
# split the test_dataset dataframe to 50% test and 50% validate. 
# (this will divide the dataset into 60% train, 20% validate and 20% test )
test_dataset, valid_dataset = train_test_split(temp_test_dataset, test_size=0.5)
print(test_dataset.shape)
print(valid_dataset.shape)
```

```python
print(f"Display the datatype of the test_dataset: {type(test_dataset)}")
print(f"Train dataset    : {train_dataset.shape}")
print(f"Test dataset     : {test_dataset.shape}")
print(f"Validate dataset : {valid_dataset.shape}")
```

```python
train_labels = train_dataset.pop('Activity')
test_labels = test_dataset.pop('Activity')
valid_labels = valid_dataset.pop('Activity')
```


### SVM Classifier



```python
#create a svm classifier
model = svm.SVC(C=1, kernel="linear")

#Train the model using the training sets
model.fit(train_dataset, train_labels)

#predict the response for test dataset
y_pred = model.predict(test_dataset)
```

```python
example_batch = test_dataset[:10]
example_result = model.predict(example_batch)
print('predicted value: ')
example_result
```

### See how the training went

```python
from sklearn import metrics
y_pred = model.predict(train_dataset)
print("Activity: ",metrics.accuracy_score(train_labels, y_pred))
```

```python
y_pred = model.predict(valid_dataset)
print("Activity: ",metrics.accuracy_score(valid_labels, y_pred))
```

```python
y_pred = model.predict(test_dataset)
print("Activity: ",metrics.accuracy_score(test_labels, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

ax = plt.subplot()
predict_results = model.predict(test_dataset)

cm = confusion_matrix(predict_results, predict_results)
sns.heatmap(cm, annot=True, ax = ax);

ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix');
```
